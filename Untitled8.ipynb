```python

copy1.decimate(factor=4, strict_length=False)
copy1_filtr=copy1.copy()
copy1_filtr.filter('lowpass', freq=0.4 * sample1.stats.sampling_rate / 4.0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Now let's plot the raw and filtered data...
t = np.arange(0, sample1.stats.npts / sample1.stats.sampling_rate, sample1.stats.delta)
t_new = np.arange(0, copy1.stats.npts / copy1.stats.sampling_rate,
                 copy1.stats.delta)

plt.plot(t, sample1.data, 'k', label='Raw', alpha=0.3)
plt.plot(t, copy1_filtr.data, 'b', label='Lowpassed', alpha=0.7)
plt.plot(t_new, copy1.data, 'r', label='Lowpassed/Downsampled', alpha=0.7)
plt.xlabel('Time [s]')
plt.xlim(82, 83.5)
plt.suptitle(sample1.stats.starttime)
plt.legend()
plt.show()
```
